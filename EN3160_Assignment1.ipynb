### Question 1:

```python
im = cv.imread('Images/emma.jpg')
im = cv.cvtColor(im, cv.COLOR_BGR2RGB)

# Converting to grayscale
im_gray = cv.cvtColor(im, cv.COLOR_RGB2GRAY)

fig, ax = plt.subplots(1, 3, figsize=(15, 5))
ax[0].imshow(im_gray, cmap='gray', vmin=0, vmax=255)
ax[0].set_title('Original Image')

#creating the tranforming function
c = np.array([(100, 50), (150, 200)])

t1 = np.linspace(0, c[0,1], c[0,0] + 1 - 0).astype('uint8')
print(len(t1))
t2 = np.linspace(c[0,1] + 1, c[1,1], c[1,0] - c[0,0]).astype('uint8')
print(len(t2))
t3 = np.linspace(c[1,1] + 1, 255, 255 - c[1,0]).astype('uint8')
print(len(t3)) 
 
transform = np.concatenate((t1, t2), axis=0).astype('uint8')
transform = np.concatenate((transform, t3), axis=0).astype('uint8')
print(len(transform))

ax[2].plot(transform)
ax[2].set_xlabel(r'Input, $f(/mathbf{x})$')
ax[2].set_ylabel('Output, $\mathrm{T}[f(\mathbf{x})]$')
ax[2].set_xlim(0,255)
ax[2].set_ylim(0,255)
ax[2].set_aspect('equal') 
plt.show()

print(transform)



```
