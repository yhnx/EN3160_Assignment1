```python
im = cv.imread('Images/emma.jpg')
im = cv.cvtColor(im, cv.COLOR_BGR2RGB)

# Converting to grayscale
im_gray = cv.cvtColor(im, cv.COLOR_RGB2GRAY)

# Displaying the original image
fig, ax = plt.subplots(1, 3, figsize=(15, 5))
ax[0].imshow(im_gray, cmap='gray', vmin=0, vmax=255)
ax[0].set_title('Original Image')

#creating the tranforming function
t1 = np.linspace(0, 50, 50).astype('uint8')
t2 = np.linspace(100, 255, 100).astype('uint8')
t3 = np.linspace(150, 255, 256-150).astype('uint8')
 
tranform1 = np.concatenate((t1, t2), axis=0).astype('uint8')
tranform1 = np.concatenate((tranform1, t3), axis=0).astype('uint8')


ax[2].plot(tranform1)
ax[2].set_xlabel(r'Input, $f(/mathbf{x})$')
ax[2].set_ylabel('Output, $/mathrm{T}[f(/mathbf{x})]$')
ax[2].set_xlim(0,255)
ax[2].set_ylim(0,255)
ax[2].set_aspect('equal') 

# Applying the transformation
im_transformed = cv.LUT(im_gray, tranform1)

print(im_transformed.shape)

ax[1].imshow(im_transformed, cmap='gray', vmin=0, vmax=255)
ax[1].set_title('Transformed Image') 
plt.show()  
```

```python
im_brain = cv.imread('Images/brain_proton_density_slice.png')
im_brain = cv.cvtColor(im_brain, cv.COLOR_BGR2RGB)

# Converting to grayscale
im_brain_gray = cv.cvtColor(im_brain, cv.COLOR_RGB2GRAY)

plt.imshow(im_brain_gray, cmap='gray', vmin=0, vmax=255)



# Displaying the original image
fig, ax = plt.subplots(1, 2, figsize=(10, 5))



# Creating the transformation function for the brain image
#creating the tranforming function
t1 = np.linspace(0, 0, 100).astype('uint8')
t2 = np.linspace(0, 255, 50).astype('uint8')
t3 = np.linspace(255, 0, 50).astype('uint8')
t4 = np.linspace(0, 0, 56).astype('uint8')
 
tranform2 = np.concatenate((t1, t2), axis=0).astype('uint8')
tranform2 = np.concatenate((tranform2, t3), axis=0).astype('uint8')
tranform2 = np.concatenate((tranform2, t4), axis=0).astype('uint8')

ax[1].plot(tranform2)
ax[1].set_xlabel(r'Input, $f(\mathbf{x})$')
ax[1].set_ylabel('Output, $/mathrm{T}[f\mathbf{x})]$')
ax[1].set_title('Grey Matter Enhancement Transformation')
ax[1].set_xlim(0,255)
ax[1].set_ylim(0,255)
ax[1].set_aspect('equal') 


# Applying the transformation
brain_transformed = cv.LUT(im_brain_gray, tranform2)
ax[0].imshow(brain_transformed, cmap='gray', vmin=0, vmax=255)
ax[0].set_title('Grey Matter Enhancement')
plt.show()


# Displaying the original image
fig, ax = plt.subplots(1, 2, figsize=(10, 5))


# Creating the transformation function for the brain image
#creating the tranforming function
t1_1 = np.linspace(0, 0, 150).astype('uint8')
t2_1 = np.linspace(0, 255, 50).astype('uint8')
t3_1 = np.linspace(255, 0, 50).astype('uint8')
t4_1 = np.linspace(0, 0, 6).astype('uint8')
 
tranform3 = np.concatenate((t1_1, t2_1), axis=0).astype('uint8')
tranform3 = np.concatenate((tranform3, t3_1), axis=0).astype('uint8')
tranform3 = np.concatenate((tranform3, t4_1), axis=0).astype('uint8')

ax[1].plot(tranform3)
ax[1].set_xlabel(r'Input, $f(\mathbf{x})$')
ax[1].set_ylabel('Output, $/mathrm{T}[f\mathbf{x})]$')
ax[1].set_title('White Matter Enhancement Transformation')
ax[1].set_xlim(0,255)
ax[1].set_ylim(0,255)
ax[1].set_aspect('equal') 


# Applying the transformation
brain_transformed = cv.LUT(im_brain_gray, tranform3)
ax[0].imshow(brain_transformed, cmap='gray', vmin=0, vmax=255)
ax[0].set_title('White Matter Enhancement')
plt.show()
```

```python
img3 = cv.imread("Images\highlights_and_shadows.jpg")
img3 = cv.cvtColor(img3, cv.COLOR_BGR2RGB)

plt.imshow(img3)

#plot the image histogram for rgb channels
fig, ax = plt.subplots(1, 3, figsize=(15, 5))
ax[0].hist(img3[:, :, 0].ravel(), bins=256, color='red', alpha=0.5, label='Red Channel')
ax[1].hist(img3[:, :, 1].ravel(), bins=256, color='green', alpha=0.5, label='Green Channel')
ax[2].hist(img3[:, :, 2].ravel(), bins=256, color='blue', alpha=0.5, label='Blue Channel')
ax[0].set_title('Histogram of RGB Channels')

```

```python
# Convert RGB to L*a*b* color space
img3_lab = cv.cvtColor(img3, cv.COLOR_RGB2LAB)

# Split the image into L*a*b* color space
L, a, b = cv.split(img3_lab)

# Display L channel analysis and gamma transformation
fig, ax = plt.subplots(1, 3, figsize=(15, 5))

# Original L channel histogram
ax[0].hist(L.ravel(), bins=256,color='black', alpha=0.7)
ax[0].set_title('Original L Channel Histogram')
ax[0].set_xlabel('L Values')
ax[0].set_ylabel('Frequency')
ax[0].grid(True, alpha=0.3)

# Define gamma transformation for L channel (proper 0-100 range)
gamma = 2.3
transform_gamma = np.array([(i/256)**(gamma)*256 for i in np.arange(0, 256)]).astype('uint8')

# Plot gamma transformation function
ax[1].plot(transform_gamma)
ax[1].set_xlabel('Input L Values')
ax[1].set_ylabel('Output L Values')
ax[1].set_title(f'Gamma Correction ( = {gamma})')
ax[1].set_aspect('equal')
ax[1].grid(True, alpha=0.3)

# Apply gamma correction to L channel
L_modified = cv.LUT(L, transform_gamma)




# Merge channels and convert back to RGB
merged = cv.merge([L_modified, a, b])
img3_transformed = cv.cvtColor(merged, cv.COLOR_LAB2RGB)


# Modified L channel histogram
L_modified = L_modified.flatten()
ax[2].hist(L_modified.ravel(), bins=256,range=(0, 256), color='red', alpha=0.7)
ax[2].set_title('Gamma Corrected L Channel')
ax[2].set_xlabel('L Values (0-256)')
ax[2].set_ylabel('Frequency')
ax[2].grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

# Display original vs gamma corrected images
fig, ax = plt.subplots(1, 2, figsize=(12, 6))
ax[0].imshow(img3)
ax[0].set_title('Original Image')
ax[0].axis('off')

ax[1].imshow(img3_transformed)
ax[1].set_title(f'Gamma Corrected ( = {gamma})')
ax[1].axis('off')

plt.tight_layout()
plt.show()

# Compare RGB histograms before and after correction
fig, ax = plt.subplots(2, 3, figsize=(15, 8))

# Original RGB histograms
colors = ['red', 'green', 'blue']
channels = ['Red', 'Green', 'Blue']

for i in range(3):
    ax[0, i].hist(img3[:, :, i].ravel(), bins=256, color=colors[i], alpha=0.7)
    ax[0, i].set_title(f'Original - {channels[i]} Channel')
    ax[0, i].set_xlabel('Pixel Values')
    ax[0, i].set_ylabel('Frequency')
    ax[0, i].grid(True, alpha=0.3)

# Gamma corrected RGB histograms
for i in range(3):
    ax[1, i].hist(img3_transformed[:, :, i].ravel(), bins=256, color=colors[i], alpha=0.7)
    ax[1, i].set_title(f'Gamma Corrected - {channels[i]} Channel')
    ax[1, i].set_xlabel('Pixel Values')
    ax[1, i].set_ylabel('Frequency')
    ax[1, i].grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
```
